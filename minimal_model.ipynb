$$
\Delta \phi = -2\gamma \sin(\phi)\\
\frac{d}{dt}\phi = \omega - \frac{2\ln(2)}{\tau}\sin(\phi)
$$

Where $\gamma$ is the magnitude of the perturbation, $\omega$ is the angular frequency of the oscillator, and
$\tau$ is the doubling time.

```julia
using Plots; plotly();
using DifferentialEquations
```

```julia
ω = 2π/24
τ = 6

function rhs(X, p, t)
    _ω, _τ = p
    [_ω - 2log(2)/τ*sin(X[1])]
end

TMAX = 120.0
X0 = [0]
prob = ODEProblem(rhs, X0, (0.0, TMAX), (ω, τ))
sol = solve(prob, saveat=0.5);
```

```julia
scatter(sol.t, mod2pi.(vcat(sol.u...)))
xticks!(0:24:120)
```
